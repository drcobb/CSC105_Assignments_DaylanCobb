```python
import pandas as pd 
```

```python
df=pd.read_csv("travel_tourism_dataset.csv")
```

```python
df
```

```python
# 1. How many observations and columns are there in the original dataset?
observations, columns = df.shape
print(f"1. Number of observations: {observations}, Number of columns: {columns}")

```

```python
# 2. Group A: How many tourists are under the age of 30? Save group A in df2
df2 = df[df['Age'] < 30]
group_a_count = df2.shape[0]
print(f"2. Number of tourists under the age of 30 in Group A: {group_a_count}")
```

```python
# 3. Group B: How many tourists are under the age of 30 and stayed in a hotel? Save group B in df3
df3 = df[(df['Age'] < 30) & (df['Stay'] == 'Hotel')]
group_b_count = df3.shape[0]
print(f"3. Number of tourists under the age of 30 who stayed in a hotel in Group B: {group_b_count}")
```

```python
# 4. Group C: How many tourists are under the age of 30, stayed in a hotel, and their cost of travel is less than $6000? Save group C in df4
df4 = df[(df['Age'] < 30) & (df['Stay'] == 'Hotel') & (df['Cost of Travel(Entire Trip)'] < 6000)]
group_c_count = df4.shape[0]
print(f"4. Number of tourists under the age of 30 who stayed in a hotel and cost of travel is less than $6000 in Group C: {group_c_count}")
```

```python
# 5. From Group C, what is the name of the tourist who is from China?
chinese_tourist = df4[df4['Nationality'] == 'China'][['First Name', 'Last Name']]
print(f"5. Name of the tourist from China in Group C:\n{chinese_tourist}")
```
