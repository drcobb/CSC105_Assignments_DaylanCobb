```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic_data.csv")
```

```python
df
```

```python
class_survived = df[df['survived'] == 1].groupby('class')['survived'].count()
```

```python
average_fare_by_town = df.groupby('embark_town')['fare'].mean().reset_index()
average_fare_by_town.rename(columns={'fare': 'Average Fare'}, inplace=True)
```

```python
average_fare_by_sex_and_town = df.groupby(['embark_town', 'sex'])['fare'].mean()
```

```python
average_fare_age_by_sex_and_town = df.groupby(['embark_town', 'sex'])[['fare', 'age']].mean()
```
