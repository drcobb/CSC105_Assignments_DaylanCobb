```python
# Create a subset of df with specific columns
df2 = df[['title', 'min_salary', 'location']]
```

```python
df3 = df[df['title'] == 'Sales Associate']
```

```python
print("Dimensions of df3:", df3.shape)
```

```python
missing_salary_values = df3['min_salary'].isnull().sum()
```

```python
print("Missing salary values in df3:", missing_salary_values)
```

```python
percentage_missing_salary = (missing_salary_values / df3.shape[0]) * 100
```

```python
print(f"Percentage of missing salary values in df3: {percentage_missing_salary:.2f}%")
```

```python
df3 = df3.dropna(subset=['min_salary'])
```

```python
df4 = df3[df3['min_salary'] > 20000]
```

```python
min_salary_in_Raleigh = df4[df4['location'].str.contains('Raleigh, NC')]['min_salary'].min()
```

```python
print(f"The minimum salary of Sales Associates working in Raleigh, NC: ${min_salary_in_Raleigh:.2f}")
```
