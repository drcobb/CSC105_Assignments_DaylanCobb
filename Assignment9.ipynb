```python
import pandas as pd 
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=pd.read_csv("healthcare_data.csv")
```

```python
df
```

```python
# Visualize the distribution of patient ages using a histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Patient Ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# What is the percentage of male patients in the dataset?
percentage_male = (df['Gender'] == 'Male').sum() / len(df) * 100
print(f"Percentage of male patients: {percentage_male:.2f}%")
```

```python
# How many patients have been diagnosed with 'Cancer'?
cancer_patients = (df['Diagnosis'] == 'Cancer').sum()
print(f"Number of patients diagnosed with Cancer: {cancer_patients}")
```

```python
# What is the average treatment cost for patients aged 60 and above?
average_cost_60_and_above = df[df['Age'] >= 60]['Treatment_Cost'].mean()
print(f"Average treatment cost for patients aged 60 and above: ${average_cost_60_and_above:.2f}")
```

```python
# Group the data by 'Blood_Type' and calculate the average treatment cost for each blood type
blood_type_avg_cost = df.groupby('Blood_Type')['Treatment_Cost'].mean().sort_values(ascending=False)
print("\nAverage treatment cost for each blood type:")
print(blood_type_avg_cost)
```

```python
# Identify the blood type with the highest average treatment cost
highest_avg_cost_blood_type = blood_type_avg_cost.idxmax()
print(f"\nBlood type with the highest average treatment cost: {highest_avg_cost_blood_type}")
```

```python
# Group the data by 'Gender' and 'Diagnosis' and calculate the average age for each combination
gender_diagnosis_avg_age = df.groupby(['Gender', 'Diagnosis'])['Age'].mean().unstack()
print("\nAverage age for each combination of Gender and Diagnosis:")
print(gender_diagnosis_avg_age)
```
